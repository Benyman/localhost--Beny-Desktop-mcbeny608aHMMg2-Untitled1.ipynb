```python
ds_Zimbabwe=xr.open_dataset('ds_zimbabwe_combined.nc')
ds_Zimbabwe=ds_Zimbabwe.where(ds_Zimbabwe !=-99)
ds_Zimbabwe
```

```python
# assigning data's to variable precip
da_Zimbabwe = ds_Zimbabwe.precip

# removing all missing values
da_Zimbabwe = da_Zimbabwe.where(da_Zimbabwe !=-99)
```

```python
coords_Zimbabwe=da_Zimbabwe.sel(lon=29, lat=-19, datetime=slice('2001','2010'))
```

```python
# MONTHLY RAINFALL TOTALS
Monthly_RR = coords_Zimbabwe.groupby('datetime.month').sum('datetime')
Monthly_RR.to_series().reset_index()
```

```python
#  LONG TERM CLIMATOLOGIES 
Longterm_Climatology_RR= coords_Zimbabwe.groupby('datetime.month').mean('datetime')
Longterm_Climatology_RR.to_series().reset_index()
```

```python
# computing Annual Totals
Annual_Totals_RR = coords_Zimbabwe.resample(datetime='Y').sum('datetime')
Annual_Totals_RR.to_series().reset_index()
```

```python
# Annual Average across Madagascar
Annual_Average_RR= coords_Zimbabwe.resample(datetime='1Y').mean('datetime')
Annual_Average_RR.to_series().reset_index()
```

```python
# Annual Totals Time series plot
#Annual_Totals_RR= coords_Zimbabwe.resample(datetime='1Y').count('datetime').groupby('datetime.year').sum('datetime')
Annual_Totals_RR.plot(color = 'g',marker='o', label='PRECIPITATION')

# ENLARGING THE FIGURE SIZE
barwidth = 0.3
plt.rcParams['figure.figsize']=[10,5]

# locating the legend
plt.legend(loc='upper right', fontsize=13)

# naming the Y-axis
plt.ylabel('PRECIPITATION (mm/day)', fontsize=16, color='b')
# naming the Axis
plt.xlabel('YEARS', fontsize=16, color='b')

plt.title('ZIMBABWE ANNUAL TOTALS PRECIPITATION VARIATION 2001-2010', fontsize=16, color='b')
    
plt.savefig('Annual_Total_timeseries.png')
plt.show()
```

```python
# Annual Average
Annual_Average_RR= coords_Zimbabwe.resample(datetime='1Y').mean('datetime')
Annual_Average_RR.plot(color = 'b',marker='o', label='PRECIPITATION')

barwidth = 0.3
plt.rcParams['figure.figsize']=[10,5]

# locating the legend
plt.legend(loc='upper right', fontsize=13)

# naming the Y-axis
plt.ylabel('PRECIPITATION (mm/days)', fontsize=16, color='r')
# naming the X Axis
plt.xlabel('YEARS', fontsize=16, color='r')

plt.title('ZIMBABWE ANNUAL AVERAGE PRECIPITATION VARIATION 2001-2010', fontsize=16, color='r')
    
plt.savefig('Annual_average_timeseries.png')
plt.show()
```

```python
#computing the number of dry PER MONTH
#selecting precipitation <1mm
dry_CRR= da_Zimbabwe.where(da_Zimbabwe <1).sel(lon=29, lat=-19, datetime=slice('2001','2010'))
dry_CRR=dry_CRR.where(dry_CRR !=-99)

dry_cmon=dry_CRR.groupby('datetime.month').mean('datetime')
dry_cmon.to_series().reset_index()
```

```python
# COMPUTING THE NUMBER FOR WET (RR>1mm) PER MONTH----------------MONTHLY----------
#selecting precipitation >1mm
wet_CRR= da_Zimbabwe.where(da_Zimbabwe >=1).sel(lon=29, lat=-19, datetime=slice('2001','2010'))
wet_CRR=wet_CRR.where(wet_CRR !=-99)
wet_CRR
wet_cmon=wet_CRR.groupby('datetime.month').mean('datetime')
wet_cmon.to_series().reset_index()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import cartopy.crs as ccrs
from cartopy import feature as cf
```

```python
# SPATIAL PLOTS FOR THE NUMBER OF DRY (RR<1mm)----------------MONTHLY--------

#selecting precipitation <1mm
dry_RR= da_Zimbabwe.where(da_Zimbabwe <1)
dry_RR=dry_RR.where(dry_RR !=-99)
dry_RR
dry_mon=dry_RR.groupby('datetime.month').mean('datetime')
dry_mon.to_series().reset_index()
fig=plt.figure()
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER
#wet_mon.plot.contourf(x='lon',y='lat', col='month',col_wrap=6);
count=0
for Months in dry_mon.month:
    count+=1
    ax=fig.add_subplot(2,6,count,projection=ccrs.PlateCarree())
    dry_mon[dry_mon.month==Months].plot()
    
    #ADDING FEATURES
    ax.coastlines(resolution='10m')
    ax.add_feature(cf.BORDERS)
    ax.add_feature(cf.LAND)
    ax.add_feature(cf.OCEAN)
    ax.set_extent ([25.3,33.3,-22.5,-16.05])
    #grid=ax.gridlines()
    #grid.xlabels_bottom=True; grid.ylabel_left=True
    #grid.xformatter=LONGITUDE_FORMATTER
    #grid.yformatter=LONGITUDE_FORMATTER
    plt.savefig('WET MONTH.png')
    #plt.show()
```
